```python
import geopandas as gpd
import pandas as pd
from collections import defaultdict

# Charger uniquement les colonnes nécessaires (sans la géométrie)
columns_needed = ["ID_PARCEL", "SURF_PARC", "CODE_CULTU"]

gdf_output = gpd.read_file("output.gpkg", layer="SELECT", usecols=columns_needed)
gdf_output2 = gpd.read_file("output2.gpkg",layer="SELECT", usecols=columns_needed)
gdf_output2.columns = gdf_output2.columns.str.upper()
gdf_output.columns = gdf_output.columns.str.upper()

# Construire un dictionnaire {ID_PARCEL: CODE_CULTU} pour un accès rapide à output2
parcel_to_cultu_output2 = dict(zip(gdf_output2["ID_PARCEL"], gdf_output2["CODE_CULTU"]))

# Initialiser la matrice d'adjacence
adj_matrix = defaultdict(lambda: defaultdict(float))

# Construire la matrice en parcourant output
for _, row in gdf_output.iterrows():
    id_parcel = row["ID_PARCEL"]
    surf_parc = row["SURF_PARC"]
    cultu_output = row["CODE_CULTU"]

    if id_parcel in parcel_to_cultu_output2:
        cultu_output2 = parcel_to_cultu_output2[id_parcel]
        adj_matrix[cultu_output][cultu_output2] += surf_parc

# Convertir en DataFrame et exporter
adj_df = pd.DataFrame.from_dict(adj_matrix, orient="index").fillna(0)
adj_df.to_csv("matrice_adjacence.csv")
```

```python
!sudo apt install libgdal-dev gdal-bin
!export CPLUS_INCLUDE_PATH=/usr/include/gdal
!export C_INCLUDE_PATH=/usr/include/gdal

cppcode = r"""

#include <iostream>
#include <unordered_map>
#include <vector>
#include <map>
#include <utility>  // Pour std::pair
#include <string>

// Structure représentant les données d'une parcelle
struct ParcelData {
    double surface;
    std::string code_group;
    std::string parcel_type;

    ParcelData(double s, const std::string& cg, const std::string& pt)
        : surface(s), code_group(cg), parcel_type(pt) {}
};

// Définition de la fonction de hachage pour std::pair
namespace std {
    template <typename T, typename U>
    struct hash<std::pair<T, U>> {
        size_t operator()(const std::pair<T, U>& p) const {
            size_t h1 = hash<T>()(p.first);
            size_t h2 = hash<U>()(p.second);
            return h1 ^ (h2 << 1);  // Combinaison des deux hachages
        }
    };
}

int main() {
    // Initialisation des grilles (utilisation de unordered_map sans boost)
    std::unordered_map<std::pair<int, int>, std::vector<ParcelData>> grid1, grid2;

    // Exemple de données dans grid1 et grid2 (à adapter selon votre logique)
    grid1[std::make_pair(0, 0)].emplace_back(100.0, "A", "old");
    grid2[std::make_pair(0, 0)].emplace_back(150.0, "B", "new");

    // Comparaison des parcelles entre grid1 et grid2
    std::map<std::pair<std::string, std::string>, double> adj;

    // Comparaison des parcelles dans grid1
    for (const auto& [cell, parcels1] : grid1) {
        for (const auto& parcel1 : parcels1) {
            // Recherche de la même cellule dans grid2
            auto it = grid2.find(cell);
            if (it != grid2.end()) {
                // Traitement des correspondances
                for (const auto& parcel2 : it->second) {
                    if (parcel1.code_group == parcel2.code_group) {
                        adj[{parcel1.code_group, parcel2.code_group}] += parcel1.surface;
                    }
                }
            } else {
                // Traitement si la cellule n'est pas trouvée dans grid2
                adj[{"0", parcel1.code_group}] += parcel1.surface;
            }
        }
    }

    // Comparaison des parcelles dans grid2
    for (const auto& [cell, parcels2] : grid2) {
        for (const auto& parcel2 : parcels2) {
            // Recherche de la même cellule dans grid1
            auto it = grid1.find(cell);
            if (it == grid1.end()) {
                // Traitement si la cellule n'est pas trouvée dans grid1
                adj[{parcel2.code_group, "0"}] += parcel2.surface;
            }
        }
    }

    // Affichage des résultats
    for (const auto& [key, value] : adj) {
        std::cout << key.first << " - " << key.second << ": " << value << std::endl;
    }

    return 0;
}


"""

with open('code.cpp', 'w') as file:
  file.write(cppcode)

!g++ code.cpp -O2 -std=c++17 -lgdal -o exe
!./exe
```

```python
#code principal et fonctionnel
!sudo apt install libgdal-dev gdal-bin
!export CPLUS_INCLUDE_PATH=/usr/include/gdal
!export C_INCLUDE_PATH=/usr/include/gdal

cppcode = r"""

#include <iostream>
#include <vector>
#include <string>
#include <fstream>
#include <map>
#include <set>
#include <unordered_map>
#include <cmath>
#include <tuple>
#include <gdal/gdal_priv.h>
#include <gdal/ogrsf_frmts.h>
#include <boost/functional/hash.hpp>
#include <iomanip>  // Pour contrôler le format d'affichage des nombres

// 📌 Définition de la grille
const double X_MIN = 100167.0, X_MAX = 1242140.0;
const double Y_MIN = 6049690.0, Y_MAX = 7108930.0;
const int GRID_SIZE = 1000;  // 1000x1000 cellules

// 📌 Calcul de l'indice de la cellule pour un point (x, y)
std::pair<int, int> get_grid_cell(double x, double y) {
    int i = (int)((x - X_MIN) / (X_MAX - X_MIN) * GRID_SIZE);
    int j = (int)((y - Y_MIN) / (Y_MAX - Y_MIN) * GRID_SIZE);
    return {std::max(0, std::min(i, GRID_SIZE - 1)), std::max(0, std::min(j, GRID_SIZE - 1))};
}

int main() {
    GDALAllRegister();

    // 📌 Stockage des parcelles dans une grille
    using ParcelData = std::tuple<double, std::string, std::string, OGRPoint>;
    std::unordered_map<std::pair<int, int>, std::vector<ParcelData>, boost::hash<std::pair<int, int>>> grid1, grid2;

    // 📌 Lire le premier fichier (ancienne version des parcelles)
    GDALDataset* poDataset1 = (GDALDataset*)GDALOpenEx("parcelles.gpkg", GDAL_OF_VECTOR, NULL, NULL, NULL);
    OGRLayer* poLayer1 = poDataset1->GetLayer(0);

    OGRFeature* poFeature;
    while ((poFeature = poLayer1->GetNextFeature()) != NULL) {
        double sp = poFeature->GetFieldAsDouble("SURF_PARC");
        std::string cg = poFeature->GetFieldAsString("CODE_CULTU");
        OGRGeometry* geom = poFeature->GetGeometryRef();
        OGRPoint pt;
        geom->Centroid(&pt);

        auto cell = get_grid_cell(pt.getX(), pt.getY());
        grid1[cell].emplace_back(sp, cg, "old", pt);

        OGRFeature::DestroyFeature(poFeature);
    }
    GDALClose(poDataset1);

    // 📌 Lire le deuxième fichier (nouvelle version des parcelles)
    GDALDataset* poDataset2 = (GDALDataset*)GDALOpenEx("parcelles2.gpkg", GDAL_OF_VECTOR, NULL, NULL, NULL);
    OGRLayer* poLayer2 = poDataset2->GetLayer(0);

    while ((poFeature = poLayer2->GetNextFeature()) != NULL) {
        double sp = poFeature->GetFieldAsDouble("SURF_PARC");
        std::string cg = poFeature->GetFieldAsString("CODE_CULTU");
        OGRGeometry* geom = poFeature->GetGeometryRef();
        OGRPoint pt;
        geom->Centroid(&pt);

        auto cell = get_grid_cell(pt.getX(), pt.getY());
        grid2[cell].emplace_back(sp, cg, "new", pt);

        OGRFeature::DestroyFeature(poFeature);
    }
    GDALClose(poDataset2);

    // 📌 Construction de la matrice d'adjacence
    std::map<std::string, double> adjMatrix;  // Stockage de la matrice sous forme de flat map
    std::set<std::string> cgs;  // Ensemble des groupes de parcelles

    for (const auto& [cell, parcels1] : grid1) {
        auto it = grid2.find(cell);

        if (it != grid2.end()) {
            auto& parcels2 = it->second;

            // 📌 Apparier les parcelles par surface et position
            for (const auto& [sp1, cg1, _, pt1] : parcels1) {
                bool matched = false;
                for (const auto& [sp2, cg2, _, pt2] : parcels2) {
                    if (std::abs(sp1 - sp2)/sp1 < 0.05 && pt1.Distance(&pt2) < 30) {  // Tolérance de 10m
                        adjMatrix[cg1 + "," + cg2] += sp1;
                        cgs.insert(cg1);
                        cgs.insert(cg2);
                        matched = true;
                        break;
                    }
                }
                if (!matched) adjMatrix["0," + cg1] += sp1;  // Parcelle disparue
            }

            for (const auto& [sp2, cg2, _, pt2] : parcels2) {
                bool matched = false;
                for (const auto& [sp1, cg1, _, pt1] : parcels1) {
                    if (std::abs(sp1 - sp2) < 1e-6 && pt1.Distance(&pt2) < 10) {
                        matched = true;
                        break;
                    }
                }
                if (!matched) adjMatrix[cg2 + ",0"] += sp2;  // Parcelle apparue
            }
        } else {
            for (const auto& [sp1, cg1, _, pt1] : parcels1)
                adjMatrix["0," + cg1] += sp1;  // Toutes les parcelles de grid1[cell] ont disparu
        }
    }

    // 📌 Vérifier les nouvelles parcelles non appariées
    for (const auto& [cell, parcels2] : grid2) {
        if (grid1.find(cell) == grid1.end()) {
            for (const auto& [sp2, cg2, _, pt2] : parcels2)
                adjMatrix[cg2 + ",0"] += sp2;  // Toutes les parcelles de grid2[cell] sont nouvelles
        }
    }

    // 📌 Ajouter "0" pour les groupes non trouvés dans la matrice
    cgs.insert("0");

    // 📌 Création de la matrice d'adjacence et écriture dans un fichier CSV
    std::ofstream outfile("adjacence.csv");
    if (!outfile) {
        std::cerr << "Erreur d'ouverture du fichier adjacence.csv\n";
        return 1;
    }

    // 🚨 Ajouter les en-têtes du CSV
    outfile << ";";
    for (const auto& cg : cgs) {
        outfile << cg << ";";
    }
    outfile << "\n";

    // 🚨 Remplir les lignes de la matrice d'adjacence avec des nombres non formatés en scientifique
    outfile << std::fixed << std::setprecision(6);  // Utiliser un format à 6 décimales
    for (const auto& cg1 : cgs) {
        outfile << cg1 << ";";
        for (const auto& cg2 : cgs) {
            // Écriture des flux entre les groupes
            outfile << adjMatrix[cg1 + "," + cg2] << ";";
        }
        outfile << "\n";
    }

    outfile.close();
    std::cout << "Matrice d'adjacence écrite dans adjacence.csv\n";
    return 0;
}


"""

with open('code.cpp', 'w') as file:
  file.write(cppcode)

!g++ code.cpp -O2 -std=c++17 -lgdal -o exe
!./exe
```

```python
# Code de Kilian
!sudo apt install libgdal-dev gdal-bin
!export CPLUS_INCLUDE_PATH=/usr/include/gdal
!export C_INCLUDE_PATH=/usr/include/gdal

cppcode = r"""

#include <iostream>
#include <vector>
#include <string>
#include <unordered_map>
#include <cstdlib>
#include <fstream>
#include <gdal/ogrsf_frmts.h>

// Fonction de hachage personnalisée pour std::pair<string, string>
struct pair_hash {
    template <class T1, class T2>
    std::size_t operator ()(const std::pair<T1, T2>& p) const {
        auto h1 = std::hash<T1>{}(p.first);
        auto h2 = std::hash<T2>{}(p.second);
        return h1 ^ h2; // Combine les deux hachages
    }
};

using namespace std;

// Fonction pour récupérer les catégories uniques depuis le fichier GeoPackage
unordered_map<string, int> getCategoriesFromGeoPackage(const char* filePath) {
    // Ouverture du fichier GeoPackage
    GDALAllRegister();
    OGRDataSource* poDS = (OGRDataSource*) OGROpen(filePath, FALSE, NULL);
    if (poDS == NULL) {
        cerr << "Erreur lors de l'ouverture du fichier GeoPackage" << endl;
        exit(1);
    }

    // Récupérer la couche contenant les parcelles
    OGRLayer* poLayer = poDS->GetLayerByName("PARCELLES_GRAPHIQUES");
    if (poLayer == NULL) {
        cerr << "Erreur : la couche 'PARCELLES_GRAPHIQUES' n'a pas été trouvée." << endl;
        exit(1);
    }

    // Créer une map pour stocker les catégories et leurs comptages
    unordered_map<string, int> categoryCount;

    // Itérer sur chaque entité dans la couche
    OGRFeature* poFeature;
    poLayer->ResetReading();
    while ((poFeature = poLayer->GetNextFeature()) != NULL) {
        // Extraire le code de culture (CODE_CULTU) depuis la feature
        const char* codeCultu = poFeature->GetFieldAsString("CODE_CULTU");
        if (codeCultu != nullptr) {
            categoryCount[codeCultu]++; // Incrémenter le compteur pour cette catégorie
        }
        OGRFeature::DestroyFeature(poFeature);
    }

    // Fermer la datasource
    GDALClose(poDS);

    return categoryCount;
}

int main() {
    // Fichier GeoPackage
    const char* gpkgFile = "parcelles.gpkg";

    // Extraire les catégories depuis le fichier GeoPackage
    unordered_map<string, int> categoryCount = getCategoriesFromGeoPackage(gpkgFile);

    // Afficher les catégories trouvées
    cout << "Catégories trouvées dans le fichier GeoPackage :\n";
    for (const auto& entry : categoryCount) {
        cout << "Code de culture: " << entry.first << ", Nombre d'occurrences: " << entry.second << endl;
    }

    // Nombre de points à générer (exemple)
    const int NB_POINTS = 1000;
    vector<string> points;

    // Assigner des catégories aux points aléatoirement (pour exemple)
    srand(time(0));
    for (int i = 0; i < NB_POINTS; i++) {
        // Choisir une catégorie aléatoire parmi celles extraites
        auto it = categoryCount.begin();
        advance(it, rand() % categoryCount.size()); // Sélectionner une catégorie au hasard
        string category = it->first;
        points.push_back(category);
    }

    // Calcul des probabilités P[R1=r1|R0=r0] à partir des catégories
    unordered_map<pair<string, string>, double, pair_hash> proba;
    for (const auto& r0 : categoryCount) {
        int n = 0;
        unordered_map<string, int> subCount;
        for (const auto& r1 : categoryCount) {
            subCount[r1.first] = 0;
        }

        for (const string& point : points) {
            if (point == r0.first) {
                n++;
                subCount[point]++;
            }
        }

        for (const auto& r1 : categoryCount) {
            proba[make_pair(r0.first, r1.first)] = (double)subCount[r1.first] / n;
        }
    }

    // Affichage des résultats dans un fichier CSV
    ofstream outFile("result.csv");
    outFile << "CATEGORY,Category1,Category2\n"; // Changer avec des catégories réelles
    for (const auto& r0 : categoryCount) {
        outFile << r0.first << ",";
        for (const auto& r1 : categoryCount) {
            outFile << proba[make_pair(r0.first, r1.first)] << ",";
        }
        outFile << "\n";
    }
    outFile.close();

    cout << "Résultats enregistrés dans result.csv" << endl;

    return 0;
}



"""

with open('code.cpp', 'w') as file:
  file.write(cppcode)

!g++ code.cpp -O2 -std=c++17 -lgdal -o exe
!./exe
```

```python
#code test - appariement par inclusion centroide
!sudo apt install libgdal-dev gdal-bin
!export CPLUS_INCLUDE_PATH=/usr/include/gdal
!export C_INCLUDE_PATH=/usr/include/gdal

cppcode = r"""

#include <iostream>
#include <vector>
#include <string>
#include <fstream>
#include <map>
#include <set>
#include <unordered_map>
#include <cmath>
#include <tuple>
#include <gdal/gdal_priv.h>
#include <gdal/ogrsf_frmts.h>
#include <iomanip>
#include <boost/functional/hash.hpp>

// 📌 Définition de la grille
const double X_MIN = 100167.0, X_MAX = 1242140.0;
const double Y_MIN = 6049690.0, Y_MAX = 7108930.0;
const int GRID_SIZE = 1000;  // 1000x1000 cellules

// 📌 Calcul de l'indice de la cellule pour un point (x, y)
std::pair<int, int> get_grid_cell(double x, double y) {
    int i = (int)((x - X_MIN) / (X_MAX - X_MIN) * GRID_SIZE);
    int j = (int)((y - Y_MIN) / (Y_MAX - Y_MIN) * GRID_SIZE);
    return {std::max(0, std::min(i, GRID_SIZE - 1)), std::max(0, std::min(j, GRID_SIZE - 1))};
}

int main() {
    GDALAllRegister();

    std::map<std::string, std::map<std::string, double>> adjMatrix;
    std::set<std::string> cgs;

    GDALDataset* poDataset1 = (GDALDataset*)GDALOpenEx("parcelles.gpkg", GDAL_OF_VECTOR, NULL, NULL, NULL);
    GDALDataset* poDataset2 = (GDALDataset*)GDALOpenEx("parcelles2.gpkg", GDAL_OF_VECTOR, NULL, NULL, NULL);

    if (!poDataset1 || !poDataset2) {
        std::cerr << "Erreur d'ouverture des fichiers." << std::endl;
        return 1;
    }

    OGRLayer* poLayer1 = poDataset1->GetLayer(0);
    OGRLayer* poLayer2 = poDataset2->GetLayer(0);

    // 📌 Stockage des parcelles de `parcelles2.gpkg` dans une grille
    std::unordered_map<std::pair<int, int>, std::vector<std::pair<std::string, OGRGeometry*>>, boost::hash<std::pair<int, int>>> grid;
    OGRFeature* poFeature;

    while ((poFeature = poLayer2->GetNextFeature()) != NULL) {
        OGRGeometry* geom = poFeature->GetGeometryRef()->clone();
        std::string cg = poFeature->GetFieldAsString("CODE_CULTU");
        OGREnvelope env;
        geom->getEnvelope(&env);

        auto cell = get_grid_cell((env.MinX + env.MaxX) / 2, (env.MinY + env.MaxY) / 2);
        grid[cell].emplace_back(cg, geom);
        cgs.insert(cg);

        OGRFeature::DestroyFeature(poFeature);
    }

    while ((poFeature = poLayer1->GetNextFeature()) != NULL) {
        double sp = poFeature->GetFieldAsDouble("SURF_PARC");
        std::string cg1 = poFeature->GetFieldAsString("CODE_CULTU");
        OGRGeometry* geom = poFeature->GetGeometryRef();
        OGRPoint pt;
        geom->Centroid(&pt);

        auto cell = get_grid_cell(pt.getX(), pt.getY());
        std::string cg2 = "0";

        if (grid.find(cell) != grid.end()) {
            for (const auto& [cg, geom2] : grid[cell]) {
                if (geom2->Contains(&pt)) {
                    cg2 = cg;
                    break;
                }
            }
        }

        adjMatrix[cg1][cg2] += sp;
        cgs.insert(cg1);
        cgs.insert(cg2);
        OGRFeature::DestroyFeature(poFeature);
    }

    GDALClose(poDataset1);
    GDALClose(poDataset2);

    std::ofstream outfile("adjacence.csv");
    outfile << ";";
    for (const auto& cg : cgs) outfile << cg << ";";
    outfile << "\n";
    outfile << std::fixed << std::setprecision(6);

    for (const auto& cg1 : cgs) {
        outfile << cg1 << ";";
        for (const auto& cg2 : cgs) {
            outfile << adjMatrix[cg1][cg2] << ";";
        }
        outfile << "\n";
    }
    outfile.close();

    std::cout << "Matrice d'adjacence écrite dans adjacence.csv" << std::endl;
    return 0;
}


"""

with open('code.cpp', 'w') as file:
  file.write(cppcode)

!g++ code.cpp -O2 -std=c++17 -lgdal -o exe
!./exe
```
